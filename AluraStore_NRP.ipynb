### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
#faturamento por loja
faturamento_centro = loja["Preço"].sum()
faturamento_zonasul = loja2["Preço"].sum()
faturamento_zonaleste = loja3["Preço"].sum()
faturamento_zonanorte = loja4["Preço"].sum()

#exibir resultados
print("Faturamento Alura Store Centro: R$", faturamento_centro)
print("Faturamento Alura Store Zona Sul: R$", faturamento_zonasul)
print("Faturamento Alura Store Zona Leste: R$", faturamento_zonaleste)
print("Faturamento Alura Store Zona Norte: R$", faturamento_zonanorte)
```

```python
import matplotlib.pyplot as plt

# Nomes das lojas e seus respectivos faturamentos
lojas = ["Alura Store Centro", "Alura Store Zona Sul", "Alura Store Zona Leste", "Alura Store Zona Norte"]
faturamentos = [faturamento_centro, faturamento_zonasul, faturamento_zonaleste, faturamento_zonanorte]
cores = ['#2E8B57', '#1E90FF', '#FF8C00', '#8A2BE2']

# Criar o gráfico de barras
plt.figure(figsize=(10, 5))
plt.bar(lojas, faturamentos, color=cores)
plt.title("Faturamento por Loja")
plt.xlabel("Alura Store")
plt.ylabel("Faturamento (R$)")

# Exibir valores em cima das barras
for i, valor in enumerate(faturamentos):
   plt.text(i, valor + (valor * 0.01), f"R$ {valor:,.2f}".replace(",", "X").replace(".", ",").replace("X", "."), ha='center')


plt.tight_layout()
plt.show()
```

# 2. Vendas por Categoria


```python
# Loja 1
print("Alura Store Centro - Vendas por categoria:")
for categoria, qtd in loja['Categoria do Produto'].value_counts().items():
    print(f"- {categoria}: {qtd}")
print()
# Alura Store Zona Sul
print("Alura Store Zona Sul - Vendas por categoria:")
for categoria, qtd in loja2['Categoria do Produto'].value_counts().items():
    print(f"- {categoria}: {qtd}")
print()
# Alura Store Zona Leste
print("Alura Store Zona Leste - Vendas por categoria:")
for categoria, qtd in loja3['Categoria do Produto'].value_counts().items():
    print(f"- {categoria}: {qtd}")
print()
# Loja 4
print("Alura Store Zona Norte - Vendas por categoria:")
for categoria, qtd in loja4['Categoria do Produto'].value_counts().items():
    print(f"- {categoria}: {qtd}")
print()
```

```python
import matplotlib.pyplot as plt

# Gráfico para Alura Store Centro
categorias_centro = loja['Categoria do Produto'].value_counts()
plt.figure(figsize=(5, 5))
plt.pie(categorias_centro, labels=categorias_centro.index, autopct='%1.1f%%', startangle=140)
plt.title("Alura Store Centro - Vendas por Categoria")
plt.axis('equal')
plt.show()

# Gráfico para Alura Store Zona Sul
categorias_zs = loja2['Categoria do Produto'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(categorias_zs, labels=categorias_zs.index, autopct='%1.1f%%', startangle=140)
plt.title("Alura Store Zona Sul - Vendas por Categoria")
plt.axis('equal')
plt.show()

# Gráfico para Alura Store Zona Leste
categorias_zel = loja3['Categoria do Produto'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(categorias_zel, labels=categorias_zel.index, autopct='%1.1f%%', startangle=140)
plt.title("Alura Store Zona Leste - Vendas por Categoria")
plt.axis('equal')
plt.show()

# Gráfico para Alura Store Zona Norte
categorias_zn = loja4['Categoria do Produto'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(categorias_zn, labels=categorias_zn.index, autopct='%1.1f%%', startangle=140)
plt.title("Alura Store Zona Norte - Vendas por Categoria")
plt.axis('equal')
plt.show()
```

# 3. Média de Avaliação das Lojas

```python
# Função para exibir apenas a média das avaliações de cada loja
def exibir_media_avaliacao(loja_df, nome_loja):
    # Média da nota (arredondada)
    media = round(loja_df['Avaliação da compra'].mean(), 2)
    print(f"Média das avaliações - {nome_loja}: {media}")

# Aplicar para cada loja
exibir_media_avaliacao(loja, "Alura Store Centro")
exibir_media_avaliacao(loja2, "Alura Store Sul")
exibir_media_avaliacao(loja3, "Alura Store Leste")
exibir_media_avaliacao(loja4, "Alura Store Norte")
```

# 4. Produtos Mais e Menos Vendidos

```python
lojas = [loja, loja2, loja3, loja4]
nomes = ['Alura Store Centro', 'Alura Store Sul', 'Alura Store Leste', 'Alura Store Norte']

# Agrupando e contando registros por categoria com groupby() e .size()
tabelas = [
    l.groupby('Categoria do Produto').size().rename(n)
    for l, n in zip(lojas, nomes)
]

# Une as tabelas, preenche com zeros e converte para inteiros
tabela_final = pd.concat(tabelas, axis=1).fillna(0).astype(int)
print(tabela_final)
```

```python
# Gráfico horizontal
tabela_plot = tabela_final.T

# Criando gráfico horizontal
tabela_plot.plot(kind='barh', figsize=(12, 6), color=cores[:tabela_plot.shape[1]])

# Títulos e rótulos
plt.title('Vendas por Loja e Categoria', fontsize=14, fontweight='bold')
plt.xlabel('Quantidade de Produtos', fontsize=14)
plt.ylabel('Lojas', fontsize=14)

# Legenda posicionada abaixo para otimizar espaço
plt.legend(title='Categoria', bbox_to_anchor=(0.5, -0.2), loc='upper center', ncol=3, fontsize=10)

# Grades discretas para facilitar leitura
plt.grid(axis='x', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

# 5. Frete Médio por Loja


```python
# Função para calcular o custo médio de frete de cada loja
def calcular_custo_frete(loja_df, nome_loja):
    # Cálculo da média do custo do frete
    custo_medio_frete = round(loja_df['Frete'].mean(), 2)
    print(f"Custo Médio de Frete - {nome_loja}: R$ {custo_medio_frete}")

# Aplicar para cada loja
calcular_custo_frete(loja, "Loja 1")
calcular_custo_frete(loja2, "Loja 2")
calcular_custo_frete(loja3, "Loja 3")
calcular_custo_frete(loja4, "Loja 4")
```
